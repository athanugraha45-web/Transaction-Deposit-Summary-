```python
import pandas as pd
df = pd.read_csv("Data_Gabungan_CDM.csv", dtype=str)
```

```python
unnamed_cols = [c for c in df.columns if c.startswith('Unnamed')]
df = df.drop(columns=unnamed_cols, errors='ignore')
```

```python
df.columns = df.columns.str.strip().str.replace(" ","_").str.replace("/", "_").str.lower()
```

```python
str_cols = df.select_dtypes(include='object').columns.tolist()
for c in str_cols :
    df[c]=df[c].astype(str).str.strip()
```

```python
# parsing start datetime
df['transactionstartdatetime_parsed'] = pd.to_datetime(df['transactionstartdatetime'], errors='coerce')

# parsing end datetime (format campuran mm/dd/yyyy hh:mm dan iso)
df['transactionenddatetime_parsed'] = pd.to_datetime(df['transactionenddatetime'], errors='coerce')

# cek berapa yang gagal parse
print("Failed parse start dt:", df['transactionstartdatetime_parsed'].isna().sum())
print("Failed parse end dt:", df['transactionenddatetime_parsed'].isna().sum())

# lihat sample baris yang gagal parse (max 20)
display(df[df['transactionstartdatetime_parsed'].isna()].head(20))
display(df[df['transactionenddatetime_parsed'].isna()].head(20))
```

```python
# buat date dari parsed start datetime
df['date'] = df['transactionstartdatetime_parsed'].dt.date

# ubah ke datetime (pandas Timestamp tanpa waktu)
df['date'] = pd.to_datetime(df['date'])

# cek unik range date
print(df['date'].min(), df['date'].max())
```

```python
# cek null di kolom kunci penting
key_cols = ['transactionid', 'transactionstartdatetime_parsed', 'transactionenddatetime_parsed', 'cardholderid', 'locationid', 'transactionamount', 'date']
print(df[key_cols].isna().sum())
```

```python
# bersihkan thousand separator jika ada (mis. "10,000")
df['transactionamount_clean'] = df['transactionamount'].str.replace(',', '').str.replace(' ', '')
df['transactionamount_clean'] = pd.to_numeric(df['transactionamount_clean'], errors='coerce')

# cek berapa null atau 0
print("NaN amount:", df['transactionamount_clean'].isna().sum())
print("Zero amount:", (df['transactionamount_clean']==0).sum())
print("Negative amount:", (df['transactionamount_clean']<0).sum())

# lihat sample value outlier (sangat besar)
display(df[df['transactionamount_clean']>100000000].head())
```

```python
dup_count = df['transactionid'].duplicated().sum()
print("Duplicate TransactionID count:", dup_count)

# jika ada duplikasi, tunjuk beberapa contoh
if dup_count>0:
    display(df[df['transactionid'].duplicated(keep=False)].sort_values('transactionid').head(50))
```

```python
# pastikan format nama kolom sama
df_location['LocationID'] = df_location['LocationID'].astype(str).str.strip()

missing_loc = df[~df['LocationID'].isin(df_location['LocationID'])]
print("Transactions with missing location in lookup:", missing_loc.shape[0])
display(missing_loc[['transactionid','locationid']].drop_duplicates().head(50))
```
